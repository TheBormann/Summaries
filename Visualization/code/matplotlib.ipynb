# Matplotlib Plots

This kernel produces all plots shown in the summary

```python
import matplotlib.pyplot as plt
import numpy as np

# Additional libraries
import matplotlib.cm as cm                  # color map
from mpl_toolkits.mplot3d import Axes3D     # 3D plots

# Data sets
import seaborn as sns
tips = sns.load_dataset("tips")
```

## Layout

```python
# Vertical Layout

plt.figure(figsize=(3, 3))
plt.subplot(2, 1, 1)
plt.xticks([])
plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2,1,1)', ha='center', va='center',
        size=24, alpha=.5)

plt.subplot(2, 1, 2)
plt.xticks([])
plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2,1,2)', ha='center', va='center',
        size=24, alpha=.5)

plt.tight_layout()
plt.savefig("../img/vertical-layout.svg")
plt.show()
```

```python
# Horizontal Layout


plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
plt.xticks([])
plt.yticks([])
plt.text(0.5, 0.5, 'subplot(1,2,1)', ha='center', va='center',
        size=24, alpha=.5)

plt.subplot(1, 2, 2)
plt.xticks([])
plt.yticks([])
plt.text(0.5, 0.5, 'subplot(1,2,2)', ha='center', va='center',
        size=24, alpha=.5)

plt.tight_layout()
plt.savefig("../img/horizontal-layout.svg")
plt.show()
```

```python
# Advanced Layout

fig = plt.figure()
fig.subplots_adjust(bottom=0.025, left=0.025, top = 0.975, right=0.975)

#  subplot(nrows, ncols, index), define rows, columns of grid and where it is
plt.subplot(4, 1, 1)
plt.xticks([]), plt.yticks([])

plt.subplot(4, 1, 2)
plt.xticks([]), plt.yticks([])

plt.subplot(2, 3, 4)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 3, 5)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 3, 6)
plt.xticks([])
plt.yticks([])

plt.savefig("../img/advanced-layout.svg")
plt.show()
```

## Different Plots

```python
# Line plot

X = np.linspace(0, 2, 200)
Y = np.sin(2*np.pi*X)

fig, ax = plt.subplots(figsize=(3,3))
ax.plot(X, Y, lw=2)

ax.grid()

plt.savefig("../img/line-plot.svg")
plt.show()
```

```python
# Scatter Plot

n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)

fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(X, Y, s=10, c=T, alpha=.5)


plt.savefig("../img/scatter-plot.svg")
plt.show()
```

```python
# Bar[h] plot

fig, ax = plt.subplots(figsize=(3,3))
ax.bar(tips.day, tips.total_bill, facecolor="C1")

plt.savefig("../img/bar-plot.svg")
plt.show()
```

```python
# Histogram

np.random.seed(1)
X = 4 + np.random.normal(0,1.5,200)

fig, ax = plt.subplots(figsize=(3,3))
ax.hist(X, bins=8, facecolor="C2", linewidth=0.5, edgecolor="white")

plt.savefig("../img/hist-plot.svg")
plt.show()
```

```python
# Boxplot

np.random.seed(10)

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]

fig, ax = plt.subplots(figsize=(3,3))
ax.boxplot(data)

plt.savefig("../img/box-plot.svg")
plt.show()
```

```python
# Violin Plot

np.random.seed(10)
D = np.random.normal((3,5,4), (0.75, 1.00, 0.75), (200,3))

fig, ax = plt.subplots(figsize=(3,3))

VP = ax.violinplot(D, [2,4,6], widths=1.5)

plt.savefig("../img/violin-plot.svg")
plt.show()
```

```python
# Contour Plot

np.random.seed(1)
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2. + X**5 + Y**3)*np.exp(-X**2-Y**2)
Z = Z - Z.min()

fig, ax = plt.subplots(figsize=(3,3))
ax.contour(X, Y, Z)

plt.savefig("../img/contour-plot.svg")
plt.show()
```

```python
# Hex Plot
np.random.seed(1)
X = np.random.uniform(1.5,6.5,100)
Y = np.random.uniform(1.5,6.5,100)
C = np.random.uniform(0,1,10000)

fig, ax = plt.subplots(figsize=(3,3))

ax.hexbin(X,Y,C, gridsize=8)


plt.savefig("../img/hex-plot.svg")
plt.show()
```

```python
# Step Plot

X = np.linspace(0, 10, 16)
Y = 4+2*np.sin(2*X)

fig, ax = plt.subplots(figsize=(3,3))
ax.step(X,Y, linewidth=0.75)

plt.savefig("../img/step-plot.svg")
plt.show()
```

```python
# Error plot

np.random.seed(1)
X = [2,4,6]
Y = [4,5,4]
E = np.random.uniform(0.5, 1.5, 3)

fig, ax = plt.subplots(figsize=(3,3))

ax.errorbar(X, Y, E, capsize=3)

plt.savefig("../img/error-plot.svg")
plt.show()
```

```python
# Barbs Plot

np.random.seed(1)
X = [[2,4,6]]
Y = [[1.5,3,2]]
U = -np.ones((1,3))*0
V = -np.ones((1,3))*np.linspace(50,100,3)

fig, ax = plt.subplots(figsize=(3,3))
ax.barbs(X,Y,U,V)
ax.set_xlim([0,8])
ax.set_ylim([0,8])

plt.savefig("../img/barbs-plot.svg")
plt.show()
```

```python
# Event Plot

np.random.seed(1)
X = [2,4,6]
D = np.random.gamma(4, size=(3, 50))

fig, ax = plt.subplots(figsize=(3,3))

ax.eventplot(D, lineoffsets=X)

plt.savefig("../img/event-plot.svg")
plt.show()
```

```python
# XCorr Plot

np.random.seed(3)
Y = np.random.uniform(-4, 4, 250)
X = np.random.uniform(-4, 4, 250)

fig, ax = plt.subplots(figsize=(3,3))

ax.xcorr(X,Y, usevlines=True)

plt.savefig("../img/xcorr-plot.svg")
plt.show()
```

```python
# Imshow Plot

delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig, ax = plt.subplots(figsize=(3,3))
im = ax.imshow(Z, interpolation='bilinear', cmap=cm.RdYlGn,
               origin='lower', extent=[-3, 3, -3, 3],
               vmax=abs(Z).max(), vmin=-abs(Z).max())

fig.savefig("../img/imshow.svg")
plt.show()

```

```python
# 3D Plot

fig = plt.figure(figsize=(3,3))
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2, 2)

fig.savefig("../img/3D-plot.svg")
plt.show()
```

```python
# Pie Chart

X = 0.5,0.125,0.125,0.25
Y = ["Apple", "Google", "Microsoft", "Huawei" ]

fig, ax = plt.subplots(figsize=(3,3))
ax.pie(X, labels=Y, radius=1, center=(1,1), startangle=90)

fig.savefig("../img/pie-plot.svg")
plt.show()
```

```python
# Text

fig, ax = plt.subplots(figsize=(3,3))
ax.text(0.5,0.5,"TEXT", size=20, weight="bold",
        ha="center", va="center", rotation=25)

fig.savefig("../img/text-plot.svg")
plt.show()
```

```python
# Quiver

np.random.seed(1)
T = np.linspace(0, 2*np.pi, 8)
X, Y = 4 + 1*np.cos(T), 4 + 1*np.sin(T)
U, V = 1.5*np.cos(T), 1.5*np.sin(T)

fig, ax = plt.subplots(figsize=(3,3))
ax.quiver(X, Y, U, V, scale=5, width=0.02)
ax.set_xlim(0, 8), ax.set_xticks(np.arange(1,8))
ax.set_ylim(0, 8), ax.set_yticks(np.arange(1,8))

fig.savefig("../img/quiver-plot.svg")
plt.show()
```

```python
# Fill Plot

np.random.seed(1)
X = np.linspace(0, 8, 16)
Y1 = 3 + 4*X/8 + np.random.uniform(0.0, 0.5, len(X))
Y2 = 1 + 2*X/8 + np.random.uniform(0.0, 0.5, len(X))

fig, ax = plt.subplots(figsize=(3,3))
ax.fill_between(X, Y1, Y2, color="C1", alpha=.5, linewidth=0)
ax.plot(X, (Y1+Y2)/2, color="C1", linewidth=0.5)

fig.savefig("../img/fill-plot.svg")
plt.show()
```
