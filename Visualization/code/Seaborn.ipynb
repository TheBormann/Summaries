# Seaborn Plots

This kernel produces all plots shown in the summary

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset("tips")
penguins = sns.load_dataset("penguins")

# Seaborn saves style setting for following plots as well
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(3,3)})

# Scale of Axis
d = 0.01

# correlation of tips
tips_mx = tips.corr()

# custom Palette
myPalette= ['#78C850',  # Grass
            '#F08030',  # Fire
            '#6890F0',  # Water
            '#A8B820',  # Bug
            '#A8A878',  # Normal
            '#A040A0',  # Poison
            '#F8D030',  # Electric
            '#E0C068',  # Ground
            '#EE99AC',  # Fairy
            '#C03028',  # Fighting
            '#F85888',  # Psychic
            '#B8A038',  # Rock
            '#705898',  # Ghost
            '#98D8D8',  # Ice
            '#7038F8',  # Dragon
           ]
```

```python
# Distplot plot
# -----------------------------------------------------------------------------
np.random.seed(0)
x = np.random.randn(100)

plt.figure(figsize=(3,3))
ax = sns.distplot(x)


plt.savefig('../img/seaborn_dist.svg')
plt.show()
```

```python
# Displot plot
# -----------------------------------------------------------------------------
np.random.seed(0)
x = np.random.randn(100)

plt.figure(figsize=(3,3))
ax = sns.displot(x, kind="hist")


plt.savefig('../img/seaborn_dis.svg')
plt.show()
```

```python
# Histogram plot
# -----------------------------------------------------------------------------
np.random.seed(0)
x = np.random.randn(100)

plt.figure(figsize=(3,3))
ax = sns.histplot(x, kde=True)


plt.savefig('../img/seaborn_hist.svg')
plt.show()
```

```python
# Relplot plot
# -----------------------------------------------------------------------------
np.random.seed(0)
x = np.random.randn(100)

plt.figure(figsize=(3,3))
ax = sns.relplot(data=tips, x="total_bill", y="tip", hue="day")


plt.savefig('../img/seaborn_rel.svg')
plt.show()
```

```python
# Joint plot
# -----------------------------------------------------------------------------

plt.figure(figsize=(3,3))
ax = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", height=3, kind="hex")

plt.savefig('../img/seaborn_joint.svg')
plt.show()
```

```python
# Kde plot
# -----------------------------------------------------------------------------
np.random.seed(1)
X = 4 + np.random.normal(0,1.5,200)

plt.figure(figsize=(3,3))
sns.kdeplot(X)

plt.tight_layout()
plt.savefig('../img/seaborn_kde.svg')
plt.show()
```

```python
# pairplot plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
ax = sns.pairplot(penguins,
        x_vars=["bill_length_mm", "bill_depth_mm"],
        y_vars=["bill_length_mm", "bill_depth_mm"],
        height=2, plot_kws=dict(marker="+", linewidth=0.2))


plt.tight_layout()
plt.savefig('../img/seaborn_pairplot.svg')
plt.show()
```

```python
# rug plot with Kde plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(3,3))
ax = sns.rugplot(X, height=.08)

plt.tight_layout()
plt.savefig('../img/seaborn_rugplot.svg')
plt.show()
```

```python
# Bar plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
ax = sns.barplot(x="day", y="total_bill", data=tips)

plt.tight_layout()
plt.savefig('../img/seaborn_bar.svg')
plt.show()
```

```python
# Count plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
ax = sns.countplot(x="day", data=tips)

plt.tight_layout()
plt.savefig('../img/seaborn_count.svg')
plt.show()
```

```python
# Box plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
ax = sns.boxplot(x="day", y="total_bill", data=tips)

plt.tight_layout()
plt.savefig('../img/seaborn_boxplot.svg')
plt.show()
```

```python
# violin plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
ax = sns.violinplot(x="day", y="total_bill", data=tips)

plt.tight_layout()
plt.savefig('../img/seaborn_violinplot.svg')
plt.show()
```

```python
# Strip plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
sns.stripplot(x="day", y="total_bill", data=tips)

plt.tight_layout()
plt.savefig('../img/seaborn_stripplot.svg')
plt.show()
```

```python
# swarm plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips, palette=myPalette)

plt.tight_layout()
plt.savefig('../img/seaborn_swarmplot.svg')
plt.show()
```

```python
# heatmap plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
sns.heatmap(tips_mx)

plt.savefig('../img/seaborn_heatmap.svg')
plt.show()
```

```python
# clustermap plot
# -----------------------------------------------------------------------------
sns.clustermap(tips_mx, figsize=(3,3))
plt.savefig('../img/seaborn_clustermap.svg')
plt.show()
```

```python
# lmplot plot
# -----------------------------------------------------------------------------
plt.figure(figsize=(3,3))
sns.set_context('paper', font_scale=1.4)

sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["o", "x"])

plt.savefig('../img/seaborn_lmplot.svg')
plt.show()
```

```python
# Pairgrid plot
# -----------------------------------------------------------------------------
g = sns.PairGrid(penguins, diag_sharey=False)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot)

plt.savefig('../img/seaborn_pairgrid.svg')
plt.show()
```

```python
# facetgrid plot
# -----------------------------------------------------------------------------
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.set_axis_labels("Total bill", "Tip")
g.add_legend()

plt.savefig('../img/seaborn_facetgrid.svg')
plt.show()
```
